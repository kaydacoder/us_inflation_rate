```python
import pandas as pd
import matplotlib.pyplot as plt

inflation_data = pd.read_csv('./1980_2023_US_inflation_rates.csv')
```

```python
# Plotting time series
plt.figure(figsize=(14, 7))
plt.plot(inflation_data['year'], inflation_data['inflation_rate'], label='Inflation Rate')
plt.plot(inflation_data['year'], inflation_data['gdp_growth'], label='GDP Growth')
plt.plot(inflation_data['year'], inflation_data['unemployment_rate'], label='Unemployment Rate')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('US Economic Indicators Over Time')
plt.legend()
plt.show()
```

```python
# Summarising statistics
summary_stats = inflation_data.describe()
print(summary_stats)
```

```python
# Analysing the correlation between inflation rate and gdp growth / unemployment rate
corr_inflation_gdp = inflation_data['inflation_rate'].corr(inflation_data['gdp_growth'])
corr_inflation_unemployment = inflation_data['inflation_rate'].corr(inflation_data['unemployment_rate'])

print(f'Correlation between Inflation Rate and GDP Growth: {corr_inflation_gdp}')
print(f'Correlation between Inflation Rate and Unemployment Rate: {corr_inflation_unemployment}')

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.scatter(inflation_data['inflation_rate'], inflation_data['gdp_growth'])
plt.xlabel('Inflation Rate')
plt.ylabel('GDP Growth')
plt.title('Inflation Rate vs GDP Growth')

plt.subplot(1, 2, 2)
plt.scatter(inflation_data['inflation_rate'], inflation_data['unemployment_rate'])
plt.xlabel('Inflation Rate')
plt.ylabel('Unemployment Rate')
plt.title('Inflation Rate vs Unemployment Rate')

plt.tight_layout()
plt.show()
```

```python
# Analysing trends
high_inflation_periods = inflation_data[inflation_data['inflation_rate'] > inflation_data['inflation_rate'].mean() + inflation_data['inflation_rate'].std()]
low_inflation_periods = inflation_data[inflation_data['inflation_rate'] < inflation_data['inflation_rate'].mean() - inflation_data['inflation_rate'].std()]

avg_gdp_growth_high_inflation = high_inflation_periods['gdp_growth'].mean()
avg_unemployment_high_inflation = high_inflation_periods['unemployment_rate'].mean()

avg_gdp_growth_low_inflation = low_inflation_periods['gdp_growth'].mean()
avg_unemployment_low_inflation = low_inflation_periods['unemployment_rate'].mean()

print(f'Average GDP Growth during High Inflation: {avg_gdp_growth_high_inflation}')
print(f'Average Unemployment Rate during High Inflation: {avg_unemployment_high_inflation}')
print(f'Average GDP Growth during Low Inflation: {avg_gdp_growth_low_inflation}')
print(f'Average Unemployment Rate during Low Inflation: {avg_unemployment_low_inflation}')
```
